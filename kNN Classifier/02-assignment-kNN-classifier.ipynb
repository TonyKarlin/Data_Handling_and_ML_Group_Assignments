## 1. Business Understanding

#### Breast Cancer Wisconsin Dataset
Build a kNN model to classify breast tumors as **Benign (B)** or **Malignant (M)** using diagnostic features.

## 2. Data Understanding

Dataset: Breast Cancer Wisconsin Diagnostic (UCI).
- Features: Numeric measurements of tumors.
- Target: `Diagnosis` (B/M)

```python
import pandas as pd
from ucimlrepo import fetch_ucirepo
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import train_test_split

# fetch dataset
breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17)

# data (as pandas dataframes)
df = breast_cancer_wisconsin_diagnostic.data.original
df.head(10)
```

Displaying the data types of the columns.

```python
df.info()
```

Displaying basic statistical information of the numerical features.


```python
df.describe(include='all')
```

## 3. Data Preparation

Splitting the dataset into two datasets:
- **X** contains all columns except `Diagnosis` and `ID`.
- **y** contains only the `Diagnosis` column.

```python
X = df.drop(columns={
    'Diagnosis',
    'ID'
})

y = df['Diagnosis']
```

Standardization is applied to ensure equal feature contribution in distance-based modeling.


```python
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

## 4. Modeling

### Train-Test Split

- Test size: 30%.

- Random state: 123.


```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size=0.3, random_state=123)
```

### Model Selection

kNN models were trained for values of k = 1 to 20.

```python
for k in range(1,21):
    model = KNeighborsClassifier(n_neighbors = k)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy_test = accuracy_score(y_test, y_pred)
    print(f'for k={k} the accuracy is: {accuracy_test:.4f}')
```



The best performance was observed for multiple values of k
`(k = 3, 9, 11, 13, 14, 15)`, all achieving the highest test accuracy of 0.9825.

We are choosing k = 3 for the final kNN model.

## 5. Evaluation

### Final Model

A kNN classifier with k = 3 was trained on the training data and evaluated on the test set.


```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

### Performance Evaluation

```python
def plot_data_to_confusion_matrix(model):
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, pos_label='M')
    recall = recall_score(y_test, y_pred, pos_label='M')
    cm = confusion_matrix(y_test, y_pred)
    cmd = ConfusionMatrixDisplay(cm, display_labels=['M','B'])
    cmd.plot()
    print(f"Accuracy : {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall   : {recall:.2f}")
```

```python
plot_data_to_confusion_matrix(model)
```

## Final Verdic

The kNN classifier with k = 3 demonstrates excellent performance on the test set.
- `98% accuracy` indicates strong overall classification performance.
- `97% precision` shows malignant predictions are highly reliable.
- `99% Recall` confirms the model’s effectiveness in correctly identifying malignant cases.

The model is now suitable for breast cancer classification tasks
